```
import cv2
import dlib
import imutils
from imutils import face_utils
from google.colab.patches import cv2_imshow
```

```
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
fa = face_utils.FaceAligner(predictor, desiredFaceWidth=256)
```

```
img = cv2.imread('dataset/positive/5.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_classifier.detectMultiScale(gray, 1.2, 6)
for (x,y,w,h) in faces:
  cv2.rectangle(img, (x,y), (x+w,y+h), (0, 255, 0), 2)
cv2_imshow(img)
```

```
img = cv2.imread('dataset/positive/5.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_classifier.detectMultiScale(gray, 1.2, 6)
if faces is ():
    print("No faces found")
for (x,y,w,h) in faces:
  faceOrig = imutils.resize(img[y:y + h, x:x + w], width=256)
  faceAligned = fa.align(img, gray, dlib.rectangle(left = x,top=y, right=x+w, bottom=y+h))
  cv2_imshow(faceOrig)
  cv2_imshow(faceAligned)
  cv2.waitKey(0)  
    
```

```
detector = dlib.get_frontal_face_detector()
```

```
img = cv2.imread('dataset/positive/pic3.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rects = detector(gray, 2)
img2 = img
for rect in rects:
  (x, y, w, h) = face_utils.rect_to_bb(rect)

  cv2.rectangle(img2, (x,y), (x+w,y+h), (0, 255, 0), 2)

cv2_imshow(img2)
```
